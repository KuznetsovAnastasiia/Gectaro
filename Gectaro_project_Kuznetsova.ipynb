Напишем функцию для вывода сведений о столбце, содержащем индексы (будем использовать её и для других датасетов).

```python
def info_about_column(datasets_name, datasets, columns_name, column):
    
    """Функция выводит общее количество строк датасета,
    количество уникальных значений конкретного столбца,
    минимальное и максимальное значения конкретного столбца"""
    
    print('Количество строк датасета', datasets_name, ' - ', datasets.shape[0], 'строк.')
    print('Количество уникальных значений столбца', columns_name, ' - ', column.nunique(), 'значений.')
    print('Значения столбца', columns_name, ': минимальное значение - ', column.min(), \
         ', максимальное значение - ', column.max())  
```

```python
info_about_column('activities', activities, '"Unnamed: 0"', activities['Unnamed: 0'])
```

Так как количество строк датасета и количество уникальных значений столбца "Unnamed: 0" совпадают, то предположение, что в этом столбце индексы подтверждается. При этом, можно рекомендовать заказчику обратить внимание на корректность формирования индексов при выгрузке данных (видно, что индексы не совпадают с количеством строк датасета).

Удалим дубликаты, перед этим напишем функцию для удаления дубликатов и вывода сведений об удалении.

```python
def drop_duplicates(table):
    
    """Функция для удаления дублирующих строк"""
    
    count_rows_original = table.shape[0]
    count_rows_start = table.shape[0]
    print('Изначальное количество строк в датасете: ', '{0:,}'.format(count_rows_original).replace(',', ' '))
    table = table.drop_duplicates().reset_index(drop=True)
    count_rows_after = table.shape[0]
    print('-' * 80)
    print('Количество удалённых строк: ', '{0:,}'.format(count_rows_start - count_rows_after).replace(',', ' '))
    print('Доля удалённых строк: ', '{0:.2%}'.format((count_rows_start - count_rows_after)/count_rows_original))
    print('-' * 80)
    print('Количество строк после удаления: ', '{0:,}'.format(count_rows_after).replace(',', ' '))
    
    return table
```

Для дальнейшего анализа данных создадим в датасетах новую колонку с группами тарифов. Сначала напишем функцию для присвоения айди тарифов групп.

```python
def create_tariffs_groups(column):
    
    '''Функция присваивает группы по айди тарифа клиента'''

    if column in [4, 7, 8, 11, 19]:
        return 'старт'
    elif column in [2, 17]:
        return 'базовый'
    elif column in [9, 20]:
        return 'бизнес'
    elif column in [3, 10, 13, 21]:
        return 'профи'
    elif column == 6:
        return 'архивный'
```

### Подготовка целевой переменной

В качестве признака ухода клиента их сервиса Gectaro возьмём последнюю (самую позднюю) дату payed_until (дата, до которой подписка компании действует) из датасета billing (платежи) и сравним её с 28.03.2024 (дата начала работы над настоящим проектом).

Для этого сформируем отдельный датасет (вспомогательный), в котором будет структура клиент - самая поздняя дата, до которой действует подписка компании, дата отсчёта (28.03.2024), разница в датах, статус клиента (ушёл или остался в сервисе).

```python
# создадим таблицу (группировка по айди клиента, оставим самую позднюю дату оплаты подписки)
billing_left_clients = billing_new[['company_id','payed_until']]\
.sort_values('payed_until')\
.groupby('company_id')\
.last()\
.reset_index()
```

```python
# добавим столбцы:
# дата начала настоящего проекта
# разница между датой оплатой подписки и датой начала проекта
billing_left_clients['projects_date'] = pd.to_datetime('2024-03-28')

# сразу переведём разницу между датами в дни, чтобы было возможно проводить расчёты
billing_left_clients['diff_date'] = (billing_left_clients['payed_until'] - billing_left_clients['projects_date']).dt.days
```

Для большего понимания выбросов построим боксплот. Перед этим напишем функцию для нахождения выбросов.

```python
def quantiles(table, column, type_format, text_sum, text_quantity):
    
    """Функция для нахождения выбросов свыше 1,5 межквартильного размаха от третьего квартиля"""
    
    Q1 = table[column].quantile(0.25)
    Q3 = table[column].quantile(0.75)
    IQR = Q3 - Q1

    print(text_sum, type_format.format(Q3  + 1.5 * IQR))
    print(text_quantity, table[table[column] > (Q3  + 1.5 * IQR)][column].count(), 'или',\
         '{0:.2%}'.format(table[table[column] > (Q3  + 1.5 * IQR)][column].count()/table.shape[0])) 
```

Напишем функцию для построения боксплотов и описания количества и границы выбросов.

```python
def create_boxplot_plus_quantiles(data,column_y,column_x,title,title_1,title_2,ylabel,\
                                  quantiles_text_total_1,quantiles_text_total_2,column_condition_1,\
                                  quantiles_text_1,column_condition_2,quantiles_text_2):
    
    """Функция строит боксплоты по признаку, а также по признаку отточности,
    и выводит сведения о количестве и границе выбросов"""

    plt.figure(figsize=(20, 10))
    plt.suptitle(title, fontsize=25)
    plt.tight_layout(h_pad= 30)    
    
    ax1 = plt.subplot(1, 2, 1)
    ax1 = sns.boxplot(y=column_y, data=data)
    plt.title(title_1, fontsize = 22)
    plt.xlabel(' ', fontsize = 10)
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.ylabel(ylabel, fontsize = 20)
    plt.grid(True)    
    
    ax2 = plt.subplot(1, 2, 2, sharey = ax1)
    ax2 = sns.boxplot(data=data, y=column_y, x=column_x)
    plt.title(title_2, fontsize = 22)
    plt.xlabel(' ', fontsize = 10)
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.ylabel(ylabel, fontsize = 20)
    plt.grid(True)
    plt.show()

    quantiles(data, column_y, '{0:,.0f}',\
              quantiles_text_total_1,\
              quantiles_text_total_2)
    display(data[column_y].describe())
    print('-'*80)
    quantiles(data[column_condition_1], column_y, '{0:,.0f}',\
              quantiles_text_1,\
              quantiles_text_total_2)
    display(data[column_condition_1][column_y].describe())
    print('-'*80)
    quantiles(data[column_condition_2], column_y, '{0:,.0f}',\
              quantiles_text_2,\
              quantiles_text_total_2)
    display(data[column_condition_2][column_y].describe())  
```

```python
create_boxplot_plus_quantiles(data_months,'month','is_left',\
                              'Распределение длительности использования клиентами сервиса Gectaro',\
                              'по всем клиентам',\
                              'по клиентам в разрезе оттока',\
                              'количество месяцев',\
                              'Выбросами являются данные о длительности использования сервиса свыше',\
                              'Количество выбросов по длительности использования сервиса',\
                              data_months.is_left == False,\
                              'Выбросами являются данные о длительности использования сервиса оставшимися клиентами свыше',\
                              data_months.is_left == True,\
                              'Выбросами являются данные о длительности использования сервиса отточными клиентами свыше')
```

- у оставшихся выбросов по продолжительности использования сервиса больше на 28%.

Также отточные клиенты не пользуются функцией сохранения информации в сервисе.

Общий портрет клиента по результатам исследования показателей активности:
- использует сервис не более 2 лет (медиана равна 6 месяцев) (похож на отточного клиента);
- выставляет в месяц до 2 счетов (медиана равна 0 счетов в месяц) (похож на отточного клиента);
- использует в месяц до 12 складов (медиана равна 3 складам) (нет разницы между отточным и оставшимся клиентами);
- не использует функцию сохранения (0 раз в месяц) (похож на отточного клиента);
- ведёт до 12 проектов в месяц (медиана равна 3 проектам) (нет разницы между отточным и оставшимся клиентами);
- до 17 контрагентов в месяц (медиана равна 3 контрагентам) (похож на отточного клиента);
- заказывает до 12 материалов в месяц (медиана равна 2 заказам) (похож на отточного клиента).

То есть, из 7 признаков типичного клиента 5 признаков описывают отточного клиента. Таким образом, для прогнозирования возможного ухода клиента из сервиса возможно использование следующих 5 признаков:
- длительность использования сервиса не превышает 24 месяцев /медиана равна 6 месяцам (у оставшихся срок больше - до 28 месяцев, медиана равна 8 месяцам);
- выставляет в месяц до 2 счетов/медиана равна 0 счетов в месяц (оставшиеся не выставляют счета (0 счетов в месяц));
- не использует функцию сохранения (0 раз в месяц) (оставшиеся используют функцию сохранения информации (до 2 раз в месяц));
- до 17 контрагентов в месяц/медиана равна 3 контрагентам (у оставшихся до контрагентов 22 в месяц, медиана равна 5 контрагентам);
- заказывает до 12 материалов в месяц/медиана равна 2 заказам (оставшиеся заказывают до 26 материалов в месяц  (медиана равна 3 заказа)).

Здесь будем основываться на составленном ранее портрете типичного клиента.

Для того, чтобы найти признаки отточных клиентов, исследовать будем следующие признаки (кодировать, чтобы также можно было применить корреляцию Пирсона):
- сколько месяцев использовал приложение Gectaro; 
- сколько всего счетов выставил в месяц;

- сколько раз сохранял данные в месяц;
- сколько проектов было в работе в месяц;
- со сколькими контрагентами работал в месяц;
- сколько раз заказывал материалы в месяц;
- являлся подписчиком тарифов "старт" и "базовый";
- в фирме не более 10 сотрудников.

Суммарные значения активностей будут необъективны: логично, что, чем дольше клиент пользуется сервисом, тем выше у него показатели (выставленные счета, используемые склады и так далее). Поэтому необходимо усреднить показатели активности. Также Заказчику будет проще понимать, что, например, клиент с большой вероятностью станет отточным, так как у него в месяц не более 1-2 контрагентов.

Начнём с того, что попробуем оценить признак: "длительность использования сервиса Gectaro отточными клиентами". Для этого создадим таблицу по количеству клиентов в разрезе месяцев.

Сначала напишем функцию для создания таблицы и гистограммы по выбранному признаку.

```python
def create_table_plus_displot(groups_column,bins_count,title_text,xlabel_text,count_rows_to_display):
    
    """Функция создаёт таблицу процентного соотношения отточных и оставшихся клиентов по выбранному признаку
    и строит гистограмму плотности распределения клиентов по признаку"""
    
    new_table = total_churn_rate.groupby([groups_column,'is_left']).agg({'company_id':'count'}).reset_index()
    
    # добавим общее количество клиентов по каждому месяцу
    total_sum = new_table.groupby(groups_column)['company_id'].sum().to_dict()
    new_table['companies_count']= new_table[groups_column].map(total_sum)

    # добавим процентное распределение между отточными и оставшимися клиентами
    new_table['ratio'] = (new_table['company_id']/new_table['companies_count'])\
    .map('{:.2%}'.format)
    
    plt.figure(figsize=(10, 6))
    sns.histplot(data=activities_total.sort_values(by=groups_column), x=groups_column, hue='is_left',  stat='density',\
                 bins=bins_count, common_norm=False, kde=True)
    plt.title(title_text)
    plt.xlabel(xlabel_text)
    ax.set(xticks=activities_total[groups_column])
    plt.ylabel('плотность')
    plt.show()
    
    
    display(new_table[groups_column].describe())
    print('Максимальный сегмент клиентов составляет ', new_table['companies_count'].max(), "клиентов(а)")
    display(new_table.head(count_rows_to_display))
```

Для расчёта количества бинов (корзин) для гистрограмм посчитаем уникальные значения для каждого признака активности.

```python
data = {'признак отточности по активности':['пользование сервисом 5 и менее месяцев',\
                                           'в месяц не более 1 счёта',\
                                           'в месяц не более 2 складов',\
                                           'не использование функции сохранения данных',\
                                           'в месяц не более 2 договоров',\
                                           'в месяц не более 2 контрагентов',\
                                           'в месяц не более 2 заказов материалов',\
                                           'пользователь тарифов "старт" или "базовый"',\
                                           'численность сотрудников не более 10 человек'],
        'размер сегмента':[76,245,151,585,152,140,121,636,638],
        'отточные клиенты':[56,189,101,398,104,98,76,424,420],
        'неотточные клиенты':[20,56,50,187,48,42,45,212,218]
       }

total_churn_rate_result = pd.DataFrame.from_dict(data)

# добавим столбцы с долями отточных и неотточных клиентов
total_churn_rate_result['доля отточных клиентов'] =\
round(total_churn_rate_result['отточные клиенты']/total_churn_rate_result['размер сегмента']*100,2)

total_churn_rate_result['доля неотточных клиентов'] =\
round(total_churn_rate_result['неотточные клиенты']/total_churn_rate_result['размер сегмента']*100,2)

# поместим столбцы по порядку
total_churn_rate_result = total_churn_rate_result[['признак отточности по активности','размер сегмента',\
                                                  'отточные клиенты','доля отточных клиентов',\
                                                  'неотточные клиенты','доля неотточных клиентов']]
```

Теперь нужно закодировать полученные признаки отточности по активности и добавить в таблицу активностей (total_churn_rate). Делаем это для того, чтобы далее к категориальным признакам можно было применить корреляцию (сначала проставим каждому клиенту группу по каждому признаку отточности по активности, затем с помощью функции dummies закодируем группы булевыми значениями).

```python
total_churn_rate.head(5)
```

```python
# для добавления групп признаков в таблицу активностей создадим её копию
churn_rate_groups = total_churn_rate.copy()

# добавим в таблицу группы признаков
# добавим признак длительности пользования сервисом (не более 5 месяцев, более 5 месяцев)
churn_rate_groups['param_months'] = ['<=5' if x > 5 else '>5' for x in churn_rate_groups['month']]

# добавим признак не более 1 счёта в месяц
churn_rate_groups['param_bills'] = ['1' if x <= 1 else '>1' for x in churn_rate_groups['billings']]

# добавим признак не более 2 складов в месяц
churn_rate_groups['param_stocks'] = ['2' if x <= 1 else '>2' for x in churn_rate_groups['stocks']]

# добавим признак не использование функции сохранения данных
churn_rate_groups['param_backups'] = ['0' if x <= 1 else '>0' for x in churn_rate_groups['backups']]

# добавим признак не более 2 договоров в месяц
churn_rate_groups['param_projects'] = ['2' if x <= 1 else '>2'\
                                        for x in churn_rate_groups['projects']]

# добавим признак не более 2 контрагентов в месяц
churn_rate_groups['param_contractors'] = ['2' if x <= 1\
                                          else '>2' for x in churn_rate_groups['contractors']]

# добавим признак не более 2 заказов материалов в месяц
churn_rate_groups['param_resources'] = ['2' if x <= 1 else '>2'\
                                       for x in churn_rate_groups['resource_orders']]

# добавим признак тарифа ("базовый или старт" и "другие")
churn_rate_groups['param_tariffs'] = ['base_start' if (x == 'базовый')|(x == 'старт') else 'not_start_base'\
                                     for x in churn_rate_groups['tariffs_group']]

# добавим признак численности сотрудников ("меньше или равно 10" или "больше 10")
churn_rate_groups['param_employers'] = ['employers<=10' if (x == '1-4')|(x == '5-10') else 'employers>10'\
                                     for x in churn_rate_groups['employers']]
```

2. Предоставить клиентам возможность подключать отдельные опции из более дорогих тарифов за дополнительную плату (также можно обеспечить выбор самими клиентами нужных им опций, не ограничивая условиями тарифа).
